```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Define the stock ticker and the date range
ticker_symbol = "AAPL" #Apple Inc.
start_date = "2025-01-01"
end_date = "2025-12-31"
```

```python
# Download the historical data 
print(f"Downloading data for {ticker_symbol}...")
stock_data = yf.download(ticker_symbol , start = start_date , end = end_date)
```

```python
# View the first few rows of data
print("\n--- First 5 rows of data ---")
print(stock_data.head())
```

```python
# Plot the "Close" price to see how it looks like
stock_data["Close"].plot(figsize = (10,5) , color = "blue")
plt.title(f"{ticker_symbol} Closing Price (2025)")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.grid(True)
```

```python
# Save the chart and show it
plt.savefig("apple_stock_2025.png")
print("\nChart saved as 'apple_stock_2025.png'!")
plt.show()
```
